### r.2020.1

For a matrix to have an orthonormal basis of eigenvectors, it must be diagonalizable in $\mathbb{R}$.
However, some orthogonal matrices are not diagonalizable in $\mathbb{R}$ but are diagonalizable in $\mathbb{C}$, the field of complex numbers.
Counterexample:
Consider the rotation matrix in $\mathbb{R}^2$ that rotates points by $90^{\circ}$ counterclockwise:
$$
A=\left[\begin{array}{cc}
0 & -1 \\
1 & 0
\end{array}\right]
$$
This matrix is orthogonal because $A^T A=A A^T=I$, where $I$ is the identity matrix.
However, the eigenvalues of $A$ are $i$ and $-i$, which are complex numbers. Since the eigenvalues are not real, there cannot exist an orthonormal basis of eigenvectors for $A$ in $\mathbb{R}^2$

For two matrices to be similar, there exists an invertible matrix $P$ such that $B=P^{-1} A P$. This means that $A$ and $B$ are diagonalizable by the same diagonal matrix, which in turn implies that they not only have the same eigenvalues but also the same eigenvectors.

While having the same eigenvalues (with algebraic multiplicities) is a necessary condition for two matrices to be similar, it is not a sufficient condition. The eigenvectors and the geometric multiplicities of the eigenvalues also play a crucial role.
Counterexample:
Consider the matrices:
$$
\begin{aligned}
& A=\left[\begin{array}{ll}
1 & 1 \\
0 & 1
\end{array}\right] \\
& B=\left[\begin{array}{ll}
1 & 0 \\
0 & 1
\end{array}\right]
\end{aligned}
$$
Both matrices have the eigenvalue 1 with algebraic multiplicity 2 . However, matrix $A$ has only one linearly independent eigenvector, while matrix $B$ has two. Thus, $A$ is not diagonalizable, but $B$ is. Since $A$ and $B$ do not have the same eigenvectors, they are not similar.
