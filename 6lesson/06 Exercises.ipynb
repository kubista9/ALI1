```python
A = Matrix([[24,-6,2,6,2],[72,51,9,-99,9],[0,-63,15,63,63],[72,15,9,-63,9],[0,63,21,-63,-27]])

vects = A.eigenvects()

l1 = vects[2][0]
l2 = vects[2][0]
l3 = vects[1][0]
l4 = vects[0][0]
l5 = vects[0][0]

D = diag(l1, l2, l3, l4, l5)

v1 = vects[2][2][0]
v2 = vects[2][2][1]
v3 = vects[1][2][0]
v4 = vects[0][2][0]
v5 = vects[0][2][1]

P = Matrix.hstack(v1, v2, v3, v4, v5)

Pinv = P**-1

A == P*D*Pinv


```
